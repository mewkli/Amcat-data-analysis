# Univariate Analysis

```python
numerical_vars = ["collegeGPA", "12percentage", "10percentage","12graduation"]
plt.figure(figsize=(16, 6))
for var in numerical_vars:
    plt.subplot(1, 4, numerical_vars.index(var) + 1)
    sns.histplot(data=data, x=var, color="pink",kde=True)
    plt.title(f"{var} Distribution")
plt.tight_layout()
plt.show()
```

collegeGPA distribution: By observing the data a maximum Collegegpa is provided around 70

12percentage distribution:By observing data most of the students score between 60 to 85 and the least percentage less than 40

10percentage Distribution:By observing distribution maximum students score 80 percentage and the least percentage less than 30

12graduation Distribution: By observing disrtribution most of the students got graduated from 2005 to 2010

```python
numerical_vars = ["openess_to_experience", "extraversion","ComputerProgramming","CivilEngg"]
plt.figure(figsize=(16, 6))
for var in numerical_vars:
    plt.subplot(1, 4, numerical_vars.index(var) + 1)
    sns.boxplot(data=data, x=var, color="pink")
    plt.title(f"{var} distribution ")
plt.tight_layout()
plt.show()
```

openess_to_experience distribution: By observing the data outliers from -2 extraversion distribution 

extraversion distribution: By observing data outliers from -2

```python
# Categorical Variables
categorical_vars = ["Degree","CollegeState"]
plt.figure(figsize=(14, 10))
for var in categorical_vars:
    plt.subplot(2, 2, categorical_vars.index(var) + 1)
    sns.countplot(data=data, x=var, color="pink", palette="Set2")
    plt.title(f"{var} Distribution")
    plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Degree Distribution:Number of students doing btech around i.e at most 3500 and the least is doing M.sc(Mtech)



CollegeState Distribution: Uttar pradesh has the maximum number of college around 800 and the least is Goa, Sikkim, Meghalaya

```python
# Categorical Variables
categorical_vars = ["Specialization","Gender"]
plt.figure(figsize=(14, 10))
for var in categorical_vars:
    plt.subplot(2, 2, categorical_vars.index(var) + 1)
    sns.countplot(data=data, x=var, palette="Set2")
    plt.title(f"{var} Distribution")
    plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Bivariate Analysis

```python
# Bivariate Analysis
numerical_vars = ["ComputerScience", "MechanicalEngg", "ElectricalEngg"]
plt.figure(figsize=(16, 6))
for i, var in enumerate(numerical_vars, 1):
    plt.subplot(1, 3, i)
    sns.scatterplot(data=data, x=var, y="Salary", color="pink")
    plt.title(f"Salary vs. {var}")
plt.tight_layout()
plt.show()
```

Salary vs ComputerScience, MechanicalEngg and ElectricalEngg: It is indicating that computer science has and ElectricalEngg have a heavy influencial Salary whereas in ElectricalEngg doesn't have a clear realation. So, discount and price doesn't effect.

```python
# Relationship between categorical and numerical columns
sns.boxplot(data=data,x=data["Specialization"],y=data["Salary"])
plt.xticks(rotation=90)
plt.title('Relationship Analysis')
```

Specialization vs Salary: Distrubtion between this relationship computer science has more outliers has compared to other specialization

```python
# Relationship between categorical and numerical columns
sns.boxplot(data=data,x=data["CollegeState"],y=data["Salary"])
plt.xticks(rotation=90)
plt.title('Relationship Analysis')
```

CollegeState v/s Salary: Distribution in uttarpradesh has highest outliers compared to other collegestate

```python
#grouping data for gender and specialization though bar plot
grouped_data = data.groupby(['Gender', 'Specialization']).size().unstack()
grouped_data.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Stacked Bar Plot of Specialization by Gender')
plt.show()
```

Gender vs Specialization: Distribution of male has more specialization compared to female

```python
# Relationship between Gender and Specialization
plt.figure(figsize=(12, 8))
sns.countplot(data=data, x='Specialization', hue='Gender')
plt.title('Countplot of Specialization by Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# Research Questions: Test the claim about Computer Science Engineering salary
cs_eng_salary = data[data['Specialization'] == 'computer science']['Salary']
mean_salary = cs_eng_salary.mean()
print(f"Mean salary of Computer Science Engineers: {mean_salary} INR")
print("Claim is True" if 250000 <= mean_salary <= 300000 else "Claim is False")
```

```python
from scipy import stats
```

```python
# Relationship between gender and specialization
gender_specialization_cross = pd.crosstab(data['Gender'], data['Specialization'])
chi2, p, _, _ = stats.chi2_contingency(gender_specialization_cross)
print(f"P-value for Chi-square test: {p}")
if p < 0.05:
    print("There is a significance")
else:
    print("There is no significance")
```
